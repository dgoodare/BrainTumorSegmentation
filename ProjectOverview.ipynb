# Semantic Segmentation of Brain Tumors from Multimodal MR Scans

This project was inspired by the BraTS (Brain Tumor Segmentation) challenge from the 2020 MICCAI conference. The challenge provides a dataset of multimodal MR images as well as four tasks to complete:
1. Segmentation of tumors
2. Predict liklihood of a patient's overall survival
3. Distinguish between pseudoprogression and true tumor recurrence
4. Evaluate algorithmic uncertainty in tumor segmentation

Here, we are only considering the first task.


Each sample in the dataset is comprised of 5 elements:
* Image data from 4 modalites, where each modality is made of 155 slices
    * Flair
    * T1
    * T1CE
    * T2
* A segmentation mask (this is what the model is trying to predict)

# U-Net
U-Net is an architecture for deep learning models that was originally designed for biomedical image segmentation tasks. But it is incredibly versatile and is used in numerous other domains. 

It is characterised by an initial down-sampling stage, where the spatial dimensions of the input image are reduced. Followed by an up-sampling stage, where the spatial dimensions are increased back to size of the input image.

![U-Net](assets/unet-architecture.png "U-Net")

Skip connections are used to concatenate the feature maps from the down-sampling stage to the corresponding stage in the upsampling path. This means that we can retain information learned from the higher resolutions that were calculated in earlier stages

### What are convolutions?
In image processing, a convlution can be thought of as applying a filter. Each pixel in the input image interacts with its neighboring pixels, applying a set of weights to each set of pixels to produce a new pixel value.
